```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution
```

```python
# Criando os pontos experimentais
x = np.linspace(0,10,500)
y = np.cos(x) + np.random.normal(0, 0.2, 500)
plt.scatter(x,y)
plt.plot(x, np.cos(x), "-r", label = "cos(x)")
plt.legend()
```

```python
def model(x, w):
  return w[0] + w[1]*x + w[2]*x**2 + w[3]*x**3 + w[4]*x**4 + w[5]*x**5
```

```python
def objetiva(w):
  y_pred = model(x,w)
  residuo = np.sqrt(sum((y - y_pred)**2)/len(y))
  return residuo
```

```python
plt.plot(x, model(x, [1.0, -0.01, 2, -0.1, 0.1, -0.01]))
plt.show()
```

```python
resultado = differential_evolution(objetiva, [(-5, 5)]*6, maxiter=2000, popsize=30)
parametros = resultado.x
```

```python
plt.scatter(x, y)
plt.plot(x, np.cos(x), "-r", label = "cos(x)")
plt.plot(x, model(x, parametros), "-k", label="resultado")
plt.legend
plt.show()
print(parametros)
```
