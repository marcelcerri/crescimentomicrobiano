```python
import numpy as np # importando o pacote Numpy
from scipy.integrate import odeint #importando a função de integração
import matplotlib.pyplot as plt #importando o Matplotlib para construção do gráfico
import os
import pandas as pd
```

```python
#Definindo a equação diferencial de resfriamento de Newton
def equation(Ca, t, mi):
   return mi*Ca
```

```python
mi = float(input("Entre com o valor de mi (1/h)"))
C_init = float(input("Entre com o valor da Concentração inicial em g/L"))
Temp_final = int(input("Entre com o valor do tempo de crescimento microbiano"))
```

```python
t = np.linspace(0,Temp_final,Temp_final*10 + 1) #criando o vetor de tempo
```

```python
Conc = odeint(equation, C_init, t, args=(mi,)) #integrando a função
```

```python
# Plotando graficamente os resultados obtidos através da integração numérica para Cxf:

## Definição das variáveis referentes ao tamanho da fonte encontrada no gráfico:

SMALL_SIZE = 14                         # Atribuição de valor numérico para a variável 'SMALL_SIZE';
MEDIUM_SIZE = 20                        # Atribuição de valor numérico para a variável 'MEDIUM_SIZE';
BIGGER_SIZE = 20                        # Atribuição de valor numérico para a variável 'BIGGER_SIZE';

## Comando para determinar o tamanho segundo o qual os textos encontrados no gráfico serão impressos na tela:

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes       # Comando para controlar tamanho padrão da fonte;
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title        # Tamanho dos números plotados nos eixos 'x' e 'y';
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels    # Tamanho da fonte utilizada para definir o título de cada eixo;
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels       # Tamanho da fonte dos números plotados no eixo 'x';
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels       # Tamanho da fonte dos números plotados no eixo 'y';
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize                   # Tamanho da fonte gráfica encontrada na legenda;
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title      # Tamanho da fonte selecionada para o título do gráfico.

## Algoritmo para plotar e imprimir os dados numéricos graficamente:

f = plt.figure()                                                     # Define a figura gerada;
_ = plt.plot(t,Conc,'b',linewidth=4)    # Variáveis 'x' e 'y', cor e tipo de marcador para traçar o perfil quando mi=0,1 h-1;
_ = plt.xlabel('Tempo (horas)',weight='bold')                 # Configuração do eixo 'x';
_ = plt.ylabel("Concentração Celular (g/L)", weight='bold')       # Configuração do eixo 'y';
_ = plt.grid(True)                                                   # Configura (presença ou ausência) a grade de marcação de fundo;
f.set_figheight(9)                                                   # Ajusta a altura da figura gerada;
f.set_figwidth(14)                                                   # Ajusta a largura da figura gerada;
f.patch.set_facecolor('white')                                       # Define a cor do fundo ao redor do gráfico;
plt.style.use('default')                                             # Aplica template 'background' original da biblioteca 'Matplotlib'.

```

```python
f.savefig('Perfil de crescimento microbiano.png')
```

```python
print("t(horas) Concentração celular (g/L)")
for x, y in list(zip(t, Conc))[::10]:
    print("%4.1f %10.2f" % (x, y))
```

```python
data = np.vstack((t, Conc.T))
data = data.T
print(data)
```
