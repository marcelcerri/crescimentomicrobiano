```python
# Integrando equação diferencial
from scipy.integrate import odeint
import numpy as np 
import matplotlib.pyplot as plt

# função a ser integrada
def crescimento (Cx, t):
    dCxdt = mi*Cx
    return dCxdt
# tempo e concentração inicial
t = np.arange (0,24,0.5)
Cx0 = 1.0
mi = 0.2

#Integração da função
Cxint = odeint(crescimento, Cx0, t)
#Construção do gráfico
plt.plot(t, Cxint, 'r-',linewidth=2, label="Concentração Celular")
plt.xlabel("tempo (horas)")
plt.ylabel("Concentração Celular")
plt.legend()
plt.show()
```

```python
#Tentando colocar o mi dentro da integração
# Integrar com diferentes mis
# Integrando equação diferencial
from scipy.integrate import odeint
import numpy as np 
import matplotlib.pyplot as plt

# função a ser integrada
def crescimento (Cx, t, mi):
    dCxdt = mi*Cx
    return dCxdt
# tempo e concentração inicial
t = np.arange (0,12,0.5)
Cx0 = 1.0

#Integração da função
mi = 0.2
Cxint02 = odeint(crescimento, Cx0, t, args=(mi,))
mi = 0.3
Cxint03 = odeint(crescimento, Cx0, t, args=(mi,))
mi = 0.4
Cxint04 = odeint(crescimento, Cx0, t, args=(mi,))
mi = 0.5
Cxint05 = odeint(crescimento, Cx0, t, args=(mi,))
#Construção do gráfico
plt.plot(t, Cxint02, 'r-',linewidth=2, label="mi = 0.2")
plt.plot(t, Cxint03, 'b-',linewidth=2, label="mi = 0.3")
plt.plot(t, Cxint04, 'g-',linewidth=2, label="mi = 0.4")
plt.plot(t, Cxint05, 'k-',linewidth=2, label="mi = 0.4")
plt.xlabel("tempo (horas)")
plt.ylabel("Concentração Celular")
plt.legend()
plt.show()
```
